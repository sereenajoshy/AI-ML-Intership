```python
df['Age_mean']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Dept_mode']=df['Department'].fillna(df['Department'].mode())
```

```python
df['Education_mode']=df['Education'].fillna(df['Education'].mode())
```

```python
df['Salary_median']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].median())
```

```python
df['City_mode']=df['City'].fillna(df['City'].mode())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_data=df.copy()
```

```python
label_encoder=LabelEncoder()
```

```python
#one hot
one_hot_encoded = pd.get_dummies(encoded_data['Gender'], prefix='Gender_en')
encoded_data = pd.concat([encoded_data, one_hot_encoded], axis=1)

#label
encoded_data['Attrition_en']=label_encoder.fit_transform(encoded_data['Attrition'])

#ordinal
ordinal_mapping = {1:'First',2:'Second',3:'Third',4:'Fourth',5:'Fivth'}
encoded_data['Performance_Rating_en']=encoded_data['Performance_Rating'].map(ordinal_mapping)

```

```python
from sklearn.preprocessing import MinMaxScaler , StandardScaler
```

```python
numeric_col=['Age','Monthly_Salary','Experience_Years']
```

```python
pf=df.copy()
```

```python
pf.dropna(subset=numeric_col, inplace=True)
```

```python
minmax_scaler= MinMaxScaler()
standard_scaler= StandardScaler()
```

```python
pf['Age_MinMax']=minmax_scaler.fit_transform(pf[['Age']])
pf['Monthly_Salary_MinMax']=minmax_scaler.fit_transform(pf[['Monthly_Salary']])
pf['Experience_Years_MinMax']=minmax_scaler.fit_transform(pf[['Experience_Years']])
```

```python
pf['Age_Standard']=standard_scaler.fit_transform(pf[['Age']])
pf['Monthly_Salary_Standard']=standard_scaler.fit_transform(pf[['Monthly_Salary']])
pf['Experience_Years_Standard']=standard_scaler.fit_transform(pf[['Experience_Years']])
```

```python
pf[['Age','Age_MinMax','Age_Standard','Monthly_Salary','Monthly_Salary_MinMax','Monthly_Salary_Standard','Experience_Years','Experience_Years_MinMax','Experience_Years_Standard']]
```
